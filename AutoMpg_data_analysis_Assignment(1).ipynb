```python
# ip/op Creation
ip=data1.drop(['mpg'],axis=1)
```

```python
ip
```

```python
op=data1['mpg']
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ip,op,test_size=0.1)
```

```python
data1.shape
```

```python
x_train.shape
```

```python
# standard scaler transform
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
alg=LinearRegression()
alg.fit(x_train,y_train)
```

```python
yp=alg.predict(x_test)
```

```python
yp
```

```python
df=pd.DataFrame({'y':list(y_test),'prediction':yp})
```

```python
x=np.array([10,307.0,130.0,3504.0,12.0,70,1]).reshape(1,-1)
sc.fit(x)
x=sc.transform(x)
alg.predict(x)
```

```python
# accuracy
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,yp)
r2=r2_score(y_test,yp)
```
