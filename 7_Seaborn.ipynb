```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
iris=pd.read_csv(r"C:\Users\asus\Downloads\iris\iris.data",header=None)
iris
```

```python
iris.columns=['SL','SW','PL','PW','flower']
iris.head()
```

```python
#Distribution plot
sns.distplot(iris.SL)
plt.show()
```

```python
sns.displot(iris.SL,fill=False,hatch='*',color='red')
plt.show()
```

```python
sns.kdeplot(iris.SL)
plt.show()
```

```python
sns.kdeplot(iris.SL,fill=True,color='g',hatch='/')
plt.show()
```

```python
sns.distplot(iris.SL[iris.flower=='Iris-setosa'])
sns.distplot(iris.SL[iris.flower=='Iris-virginica'])
sns.distplot(iris.SL[iris.flower=='Iris-versicolor'])
plt.show()
```

```python
sns.histplot(iris.SL[iris.flower=='Iris-setosa'])
sns.histplot(iris.SL[iris.flower=='Iris-virginica'])
sns.histplot(iris.SL[iris.flower=='Iris-versicolor'])
plt.show()
```

```python
sns.kdeplot(iris.SL[iris.flower=='Iris-setosa'])
sns.kdeplot(iris.SL[iris.flower=='Iris-virginica'])
sns.kdeplot(iris.SL[iris.flower=='Iris-versicolor'])
plt.show()
```

```python
sns.histplot(data=iris)
plt.show()
```

```python
sns.kdeplot(data=iris,fill=True,hatch='/')
plt.show()
```

```python
sns.boxplot(x=iris.flower,y=iris.SW,hue=iris.flower)
plt.show()
```

```python
#Violin Plot
sns.violinplot(x=iris.flower,y=iris.SW,hue=iris.flower)
plt.show()
```

```python
#Swarm Plot
sns.swarmplot(x=iris.flower,y=iris.SW,hue=iris.flower)
plt.show()
```

```python
#Srip Plot
sns.stripplot(x=iris.flower,y=iris.SW,hue=iris.flower)
plt.show()
```

```python
sns.stripplot(x=iris.flower,y=iris.SW,hue=iris.SL)
plt.show()
```

```python
#Boxen Plot
sns.boxenplot(x=iris.flower,y=iris.SW,hue=iris.flower)
plt.show()
```

```python
#Bar plot
sns.barplot(x=iris.flower,y=iris.SW,hue=iris.flower)
plt.show()
```

```python
#Point plot
sns.pointplot(x=iris.flower,y=iris.SW)
plt.show()
```

```python
sns.pointplot(x=iris.flower,y=iris.SW,hue=iris.flower)
plt.show()
```

```python
#Scatter plot
sns.scatterplot(x=iris.flower,y=iris.SW,hue=iris.flower)
plt.show()
```

```python
#Joint Plot
sns.jointplot(data=iris)
plt.show()
```

```python
#Joint Plot
sns.jointplot(data=iris,kind='kde')
plt.show()
```

```python
#Joint Plot
sns.jointplot(x=iris.SL,y=iris.SW)
plt.show()
```

```python
sns.jointplot(x=iris.SL,y=iris.SW,hue=iris.flower)
plt.show()
```

```python
#Count Plot
sns.countplot(x=iris.flower,hue=iris.flower)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x=iris.SL,hue=iris.SL,palette='cool')
plt.show()
```
