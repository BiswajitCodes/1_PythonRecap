```python
ip=data1.drop(['income'],axis=1)
```

```python
op=data1['income']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ip,op,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
alg1=LogisticRegression()
alg1.fit(x_train,y_train)
```

```python
yp=alg1.predict(x_test)
```

```python
df=pd.DataFrame({'y':list(y_test),'prediction':yp})
```

```python
df
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm=confusion_matrix(y_test,yp)
cm
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
acc=accuracy_score(y_test,yp)
recall=recall_score(y_test,yp)
f1=f1_score(y_test,yp)
prec=precision_score(y_test,yp)
```

```python
print(acc)
print(recall)
print(prec)
print(f1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
alg2=KNeighborsClassifier(n_neighbors=5)
alg2.fit(x_train,y_train)
```

```python
yp2=alg2.predict(x_test)
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
recall2=recall_score(y_test,yp2)
f12=f1_score(y_test,yp2)
prec2=precision_score(y_test,yp2)
acc2=accuracy_score(y_test,yp2)
```

```python
print(acc2)
print(recall2)
print(prec2)
print(f12)
```

```python
from sklearn.naive_bayes import GaussianNB
clf=GaussianNB()
clf.fit(x_train,y_train)
yp1=clf.predict(x_test)
```

```python
from sklearn import metrics
recall=metrics.recall_score(y_test,yp1)
accuracy=metrics.accuracy_score(y_test,yp1)
print(recall)
print(accuracy)
```

```python
from sklearn import svm
salg=svm.SVC(kernel='rbf',C=1000,gamma=0.01)
salg.fit(x_train,y_train)
yp11=salg.predict(x_test)
```

```python
from sklearn import metrics
recall=metrics.recall_score(y_test,yp11)
accuracy=metrics.accuracy_score(y_test,yp11)
print(recall)
print(accuracy)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(criterion='entropy')

clf = clf.fit(x_train,y_train)

y_pred=clf.predict(x_test)

from sklearn import metrics
accuracy2=metrics.accuracy_score(y_test,y_pred)
recall2=metrics.recall_score(y_test,y_pred)
print(recall2)
print(accuracy2)
```
