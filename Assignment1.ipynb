```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
states = ['AP', 'WB', 'OR']
workers = [100, 200,150]
non_workers = [200, 100,150]
```

```python
x = np.arange(len(states))
```

```python
plt.bar(x, workers,color='royalblue', label='Workers')
plt.bar(x, non_workers, bottom=workers,color='orange',label='Non-Workers')

plt.xlabel('states')
plt.ylabel('workers')
plt.title('States vs Workers')
plt.legend()
plt.xticks(x,states,rotation=45)
plt.show()
```

```python
labels = ['1974-75', '1978-79']
autumn = [7.83, 14.31]
winter = [85.88, 80.85]
summer = [6.29, 4.84]
```

```python
x = np.arange(len(labels))
width = 0.5
```

```python
fig, ax = plt.subplots()
ax.bar(x, autumn, width, label='Autumn Rice')
ax.bar(x, winter, width, bottom=autumn, label='Winter Rice')
bottom_summer = np.array(autumn) + np.array(winter)
ax.bar(x, summer, width, bottom=bottom_summer, label='Summer Rice')
ax.set_ylabel('Percentage')
ax.set_title('Percentage of Rice Production by Season')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```
